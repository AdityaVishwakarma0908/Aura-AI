```python
# using model llama3.2

OLLAMA_API = "http://localhost:11434/api/chat"
HEADERS = {"Content-Type": "application/json"}
MODEL = "llama3.2"

print("configuration is done ...")
```

```python
# imports required

import requests
import json
from bs4 import BeautifulSoup
from IPython.display import Markdown, display

print("imports are done ...")
```

```python
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/117.0.0.0 Safari/537.36"}

class Website:

    def __init__(self, url):

        self.url = url
        response = requests.get(url, headers=headers)
        soup = BeautifulSoup(response.content, 'html.parser')
        self.title = soup.title.string if soup.title else "No title found"
        for irrelevant in soup.body(["script", "style", "img", "input"]):
            irrelevant.decompose()
        self.text = soup.body.get_text(separator="\n", strip=True) if soup.body else "No content found"

print("website class is done ...")
```

```python
summrerizer_system_prompt = "You are an assistant that analyzes the contents of a website \
Respond in markdown."
```

```python
def user_prompt_for(website):
    user_prompt = f"You are looking at a website titled {website.title}"
    user_prompt += "\nThe contents of this website is as follows; \
please provide a short summary of this website in markdown. \
If it includes news or announcements, then summarize these too.\n\n"
    user_prompt += website.text
    return user_prompt
```

```python
def messages_for(website):
    return [
        {"role": "system", "content": summrerizer_system_prompt},
        {"role": "user", "content": user_prompt_for(website)}
    ]
```

```python
def summarize(url):

    website = Website(url)

    data = {
        "model": MODEL,
        "messages": messages_for(website),
        "stream": False
    }

    response = requests.post( 
        OLLAMA_API,
        headers = HEADERS,
        data = json.dumps(data)
    )

    result = response.json()["message"]["content"]
    return result

print("summary is ready ...")
```
