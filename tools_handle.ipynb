```python
def handle_tool_calls(message):
    responses = []
    for tool_call in message.tool_calls:
        args = json.loads(tool_call.function.arguments)
        if tool_call.function.name == "get_ticket_price":
            content = get_ticket_price(args.get('destination_city'))
        elif tool_call.function.name == "set_ticket_price":
            content = set_ticket_price(args.get("city"), args.get("price"))
                # Inside handle_tool_calls(message)
        elif tool_call.function.name == "generate_summary":
            content = generate_summary(args.get("url"))
        
        responses.append({
            "role": "tool",
            "content": content,
            "tool_call_id": tool_call.id
        })
    return responses
```
